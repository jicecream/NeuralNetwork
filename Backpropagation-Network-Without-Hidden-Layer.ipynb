```python
import numpy as np
import pandas as pd
from matplotlib  import pyplot as plt
import math
```

```python
params = pd.read_csv('params.txt', delimiter ='\t')
training = pd.read_csv('training.txt', delimiter ='\t')
```

```python
x = np.array(training['x'])
y = np.array(training['y'])
b = 1

n = x.size



```

```python

# populate training 1 to 5

training_1 = np.array(training['Training 1'])

training_2 = np.array(training['Training 2'])

training_3 = np.array(training['Training 3'])

training_4 = np.array(training['Training 4'])

training_5 = np.array(training['Training 5'])
```

```python
training_set = []
training_set = [training_1, training_2, training_3, training_4, training_5]
```

```python
weight_1 = np.array(params.iloc[0, [1,2,3]])
weight_2 = np.array(params.iloc[1, [1,2,3]])
weight_3 = np.array(params.iloc[2, [1,2,3]])
weight_4 = np.array(params.iloc[3, [1,2,3]])
weight_5 = np.array(params.iloc[4, [1,2,3]])
weight_6 = np.array(params.iloc[5, [1,2,3]])
weight_7 = np.array(params.iloc[6, [1,2,3]])
weight_8 = np.array(params.iloc[7, [1,2,3]])
weight_9 = np.array(params.iloc[8, [1,2,3]])
weight_10 = np.array(params.iloc[9, [1,2,3]])
```

```python
weight_set = []

weight_set = [weight_1 , weight_2 ,  weight_3 ,  weight_4 ,  weight_5 ,  weight_6 ,  weight_7 , 
              weight_8 ,  weight_9 ,  weight_10]
```

```python
learning_rate = float(params.iloc[0, [4]]) 
```

```python
def sigmoid(x):
    return np.where(
            x >= 0, # condition
            1 / (1 + np.exp(-x)), # For positive values
            np.exp(x) / (1 + np.exp(x)) # For negative values
    )
```

```python
def isLogBase2(n):
    if (n == 0):
        return False
    while (n != 1):
        if (n % 2 != 0):
            return False
        n = n // 2
 
    return True
    
```

```python
# for log2 error / epoch

error_list = []
error_set = []

epoch_set = []
epoch_list = []


for w in range (len(weight_set)):
    for t in range(len(training_set)):    
        
        w0 = weight_set[w][0]
        wa = weight_set[w][1]
        wb = weight_set[w][2]

        
        # forward propagation

        results = []
     
        #file = open("myFile.txt", "a") # for testing and debugging purpose
        file = open("190429166Results1.txt", "a")
        file.write("Weight set: %s" % (w+1) + " | ")
        file.write("Training set: %s" % (t+1) + " | \n")
        

        file.close()

        for epoch in range(1, 1000):
            
            z = np.add((np.add(np.dot(x, wa), np.dot(y, wb))), np.dot(b, w0))
            a = sigmoid(z)
            e = np.subtract(a, training_set[t])
          
            # find mean square error
            mse = np.sum(np.square(e))/n
                  
            if (mse< 0.000105):
                file.close()
                break
                
         
            dw = e * a * (1-a)
            w1_update = np.dot(x, dw)/n
            w2_update= np.dot(y, dw)/n
            b_update = np.sum(dw)/n

            
            wa = wa - learning_rate * w1_update
            wb = wb - learning_rate * w2_update
            w0 = w0 - learning_rate * b_update
            

            if(isLogBase2(epoch)):
                error_set.append(mse)
                epoch_set.append(epoch)
                result = "Epoch %s:|Error:%s|wa:%s|wb:%s|w0:%s" % (epoch, mse, wa,wb, w0)
#                 file = open("myFile.txt", "a") # for testing and debugging purpose
                file = open("190429166Results1.txt", "a")
                file.write(result + "\n")

        error_list.append(error_set)
        epoch_list.append(epoch_set)
        error_set = [] 
        epoch_set = []
 
        
        
        file.write("\n")
        file.close()
        
      
```

```python
index = 0

print("Results") 

for w in range (len(weight_set)):
    for t in range(len(training_set)):
                
                title = "Weight set: %s" % (w+1) + "\n" + "Training set: %s" % (t+1)
                
                x_axis = epoch_list[index]
                y_axis = error_list[index]
                
                ax = plt.axes()
                plt.plot(x_axis, y_axis)
                plt.xlabel('log2(Epoch)')
                plt.ylabel('MSE')
                plt.title(title)
                
                
                plt.xlim([0.1, max(x_axis)])
                plt.show()
                
               
                print()
                    
                index = index + 1
                
        
```
